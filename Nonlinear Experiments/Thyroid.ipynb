```python
x_train, y_train, x_test, y_test, anom_ids, anom_label, anom_label_data = get_data(
    data_dir="../thyroid+disease", train_file='ann-train.data', test_file='ann-test.data',
             normal_labels=[3], seen_anom_label=1,
             anom_label={1: 'Hyperfunction',
              2: 'Subnormal'})
```

```python
data_dir = "../Anomaly Detection/thyroid+disease"
df = pd.read_csv(f"{data_dir}/ann-train.data", header=None, delimiter=' ').dropna(axis=1)
df
```

```python
seed = 0
np.random.seed(seed)
np.random.shuffle(x_train)
np.random.seed(seed)
np.random.shuffle(y_train)
```

```python
sigma=7.
train=False  # train sigma. if NA, then False
neurons=[200]
verbose=1  # can change this to 0 to suppress verbosity during training
plot=True
shuffle=False
val_split=0.15
repeats=3
epochs=500
batchnorm=True
# lr=3e-4
lr = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.01,
    decay_steps=1000,
    decay_rate=0.96,
    staircase=True)
early_stopping = tf.keras.callbacks.EarlyStopping(patience=50, monitor='val_loss',
                                                  restore_best_weights=True)
callbacks = [early_stopping]
# callbacks=[]

results = []

for sep, activation in zip(
        ["ES", "HS"], ["b", ""]
):
    print(sep)

    results_df = test_model(
        x_train, y_train.astype(float), x_test, y_test.astype(float), x_train.shape[-1],
        anom_ids=anom_ids, anom_label=anom_label, anom_label_data=anom_label_data,
        separation=sep,
        bumped=activation,
        sigma=sigma,
        train=train,  # train sigma. if NA, then False
        neurons=neurons,
        verbose=verbose,  # can change this to 0 to suppress verbosity during training
        plot=plot,
        shuffle=shuffle,
        val_split=val_split,
        repeats=repeats,
        epochs=epochs,
        batchnorm=batchnorm,
        lr=lr,
        callbacks=callbacks
    )

    results.append(results_df)

result_comparison4 = pd.concat(results, axis=1)
result_comparison4
```
